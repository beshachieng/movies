```python
df2 = df.fillna(value=0)
df2
```

```python
# data types in the colums

df.dtypes
```

```python
# changing data types

df2['budget'] = df2['budget'].astype('int64')
df2['gross'] = df2['gross'].astype('int64')
```

```python
df2.head()
```

```python
# create correct year colum
df2['year correct'] = df2['released'].astype(str).str[-20:-16]
df2.head()
```

```python
#sorting by revenues
df2.sort_values(by=['gross'],inplace=False, ascending=False)
df2.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Checking for duplicates and droping them

df2['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# scatter plot with budget verses gross

plt.scatter(x=df2['budget'],y=df2['gross'])

plt.title('budget vs gross earnings')

plt.xlabel('budget for film')

plt.ylabel('gross earnings')

plt.show()
```

```python
#ploting budget verses gross using seaborn

sns.regplot(x='budget',y='gross',data=df2,scatter_kws={'color':'red'},line_kws={"color":'blue'})
```
